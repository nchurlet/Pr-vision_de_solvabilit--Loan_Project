En poursuivant la lecture, je lis que LabelEncoder est Ã  utiliser quand on a des donnÃ©es qui se suivent cardinalement. C'est Ã  dire qu'elles ont un rapport hiÃ©rarchique entre elles. Il faut donc bien le prendre en compte. Par exemple, si on a chien tortue chat , on obtient avec LabelEncoder 0 1 et 2. Or, chien n'est pas plus proche de tortue qu'il ne l'est de chat. Ces catÃ©gories ne sont pas cardinales. En revanche, jeune, adulte et vieux ont bien un rapport cardinal

J'en conclue que ce n'est pas un soucis quand il n'y a que 2 catÃ©gories et dans ce cas je prÃ©viligie LabelEncoder qui est moins coÃ»teux en mÃ©moire.

En effet, OneHotEncoder crÃ©e une matrice du genre pour chien : chien = 1 chat=0 tortue=0 donc pour chat : chien = 0 chat =1 tortue = 0. C'est effectivement coÃ»teux.
